```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('Soil_loss_lulc_yr.csv')
```

```python
df.head()
```

```python
#df= df.drop(df.columns[0], axis= 1)
```

```python
ax = sns.boxplot(data= df, orient='h', x="Tons.yr-1", y="LULC", hue="Fire", width=0.8)
ax.figure.set_size_inches(10,5 )
ax.set_title('Soil loss', fontsize=14)
ax
fig = ax
plt.savefig ('Soil_loss_balance.tif')
```
