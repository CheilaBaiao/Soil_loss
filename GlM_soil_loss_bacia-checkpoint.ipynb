```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from matplotlib import cm
import pandas as pd
import numpy as np
import seaborn as sns
from scipy import stats
dir(stats)
import csv
import statsmodels
import statsmodels.api as sm
import statsmodels.formula.api as smf
from matplotlib import pyplot as plt

plt.rc("figure", figsize=(16,8))
plt.rc("font", size=14)
#from matplotlib import rcParams
#rcParams['text.usetex'] = True
#rcParams['text.latex.preamble'] = r'\usepackage{amsmath}'

from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures
```

```python
df = pd.read_excel ('SDR_burned_area_bacia.xlsx') #Read excel file
df = df[df['EXPORT_BALANCE'] > 0]
df = df[df['Watershed']==43]
df = df.drop (["YEAR", "IBGE"], axis = 1)
df.info
```

```python
formula = "USLE_BALANCE~AREA+DROUGHT+ELEVATION+SLOPE+RADIATION+K_FACTOR+R_FACTOR+C_FACTOR+PRIORITY+SEVERITY+RIVERS+ROADS"#+ELEVATION*SLOPE*RADIATION+K_FACTOR*R_FACTOR+K_FACTOR*R_FACTOR*SEVERITY"
usle = smf.glm(formula=formula, data=df, family=sm.families.Gamma(link=sm.families.links.log())).fit()
print ('USLE_balance')
print('D:',round (usle.deviance,3)), print('AIC:', round(usle.aic,3)), print ('x2:', round(usle.pearson_chi2,3))
#print(usle.params)
#print(usle.pvalues)
formula = "EXPORT_BALANCE~AREA+DROUGHT+ELEVATION+SLOPE+RADIATION+K_FACTOR+R_FACTOR+C_FACTOR+PRIORITY+SEVERITY+RIVERS+ROADS"
exp = smf.glm(formula=formula, data=df, family=sm.families.Gamma(link=sm.families.links.log())).fit()
print ('Export_balance')
print('D:',  round(exp.deviance,3)), print('AIC:', round(exp.aic,3)), print ('x2:', round(exp.pearson_chi2,3))
#print(exp.params)
#print(exp.pvalues)
df = df[df['RETENTION_BALANCE'] > 0]
formula = "RETENTION_BALANCE~AREA+DROUGHT+ELEVATION+SLOPE+RADIATION+K_FACTOR+R_FACTOR+C_FACTOR+PRIORITY+ SEVERITY+RIVERS+ROADS"#+ELEVATION*SLOPE*RADIATION+K_FACTOR*R_FACTOR+K_FACTOR*R_FACTOR*SEVERITY"
ret = smf.glm(formula=formula, data=df, family=sm.families.Gamma(link=sm.families.links.log())).fit()
print ('Retention_balance')
print('D:',  round(ret.deviance,3)), print('AIC:', round(ret.aic,3)), print ('x2:', round(ret.pearson_chi2,3))
#print(ret.params)
#print(ret.pvalues)

usle = usle.summary()
u_results_as_html = usle.tables[1].as_html()
us = pd.read_html(u_results_as_html, header=0, index_col=0)[0]
us.insert(4, "Balance (Post - pre-fire)", ['Soil loss', 'Soil loss', 'Soil loss','Soil loss', 'Soil loss',  'Soil loss', 'Soil loss','Soil loss',  'Soil loss', 'Soil loss', 'Soil loss', 'Soil loss', 'Soil loss'], True)

exp = exp.summary()
x_results_as_html = exp.tables[1].as_html()
ex = pd.read_html(x_results_as_html, header=0, index_col=0)[0]
ex.insert(4, "Balance (Post - pre-fire)", ['Sediment export', 'Sediment export','Sediment export','Sediment export','Sediment export','Sediment export','Sediment export','Sediment export','Sediment export','Sediment export','Sediment export','Sediment export','Sediment export'], True)
ret = ret.summary()
r_results_as_html = ret.tables[1].as_html()
re = pd.read_html(r_results_as_html, header=0, index_col=0)[0]
re.insert(4, "Balance (Post - pre-fire)", ['Sediment retention', 'Sediment retention','Sediment retention','Sediment retention','Sediment retention','Sediment retention','Sediment retention','Sediment retention','Sediment retention','Sediment retention','Sediment retention','Sediment retention','Sediment retention'], True)
total = pd.concat([us,ex,re])
total
```

```python
import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_xticklabels():
        text = label.get_text()
        labels.append(textwrap.fill(text, width=width,
                      break_long_words=break_long_words))
    ax.set_xticklabels(labels, rotation=0)
```

```python


ax = sns.barplot(data=total, x='Variables', y='coef', hue= 'Balance (Post - pre-fire)')
threshold = 0
plt.axhline(threshold, color='black',linewidth=0.5, ls='dotted')
ax.figure.set_size_inches(12,4)
ax.set_title('(F) SSF (n=13)', fontsize=16)
#plt.xticks(fontsize=12)
ax.legend_.remove()
#plt.legend(loc='upper right')
ax.set(xlabel=None)
wrap_labels(ax, 10)
ax
fig = ax
#plt.savefig ('GLM_SSF.tif')
```
