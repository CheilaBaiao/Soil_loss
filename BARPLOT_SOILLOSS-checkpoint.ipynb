```python
import numpy  as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import cm
import pandas as pd

import seaborn as sns
from scipy.stats import kstest
from sklearn import linear_model
from sklearn.preprocessing import PolynomialFeatures

%matplotlib inline
```

```python
df = pd.read_excel('Barplot.xlsx')
```

```python
df.head()
```

```python
df = df[df['P-value'] < 0.05]
df = df[df['lulc']== 'Total']

```

```python
df.info
```

```python
import textwrap
def wrap_labels(ax, width, break_long_words=False):
    labels = []
    for label in ax.get_xticklabels():
        text = label.get_text()
        labels.append(textwrap.fill(text, width=width,
                      break_long_words=break_long_words))
    ax.set_xticklabels(labels, rotation=0)
```

```python
ax = sns.barplot(data=df, x='Variables', y='Coef', hue= 'Balance (post - pre-fire)')
threshold = 0
plt.axhline(threshold, color='black',linewidth=0.5, ls='dotted')
ax.figure.set_size_inches(12,3)
ax.set_title('Total areas(n=529)', fontsize=16)
ax.set(xlabel=None)
wrap_labels(ax, 10)
ax
fig = ax
plt.savefig ('Total.tif')
```
